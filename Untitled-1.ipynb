```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
import easygui

# Charger le modèle pré-entraîné
model_path = 'model/emotion_model.keras'  # Assure-toi que ce modèle est dans ton répertoire de travailimport cv2
import numpy as np
from tensorflow.keras.models import load_model
import easygui

# Charger le modèle pré-entraîné
model_path = 'model/emotion_model.keras'  # Assure-toi que ce modèle est dans ton répertoire de travail
emotion_model = load_model(model_path)
emotion_labels = ['Colère', 'Dégoût', 'Peur', 'Joie', 'Tristesse', 'Surprise', 'Neutre']

# Charger le détecteur de visages Haar Cascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Fonction pour détecter les visages et prédire les émotions
def detect_and_predict_emotion(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)
    
    for (x, y, w, h) in faces:
        # Extraire et prétraiter le visage
        face = gray[y:y+h, x:x+w]
        face = cv2.resize(face, (48, 48))  # Redimensionner pour correspondre à l'entrée du modèle
        face = face / 255.0  # Normaliser les pixels entre 0 et 1
        face = np.expand_dims(face, axis=0)  # Ajouter la dimension batch
        face = np.expand_dims(face, axis=-1)  # Ajouter la dimension des canaux (1 pour grayscale)

        # Prédire l'émotion
        emotion_prediction = emotion_model.predict(face, verbose=0)
        emotion_label = emotion_labels[np.argmax(emotion_prediction)]
        confidence = np.max(emotion_prediction) * 100

        # Annoter l'image avec l'émotion détectée
        cv2.putText(image, f"{emotion_label} ({confidence:.2f}%)", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
        cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)

    return image

# Menu principal
print("Choisissez une option :")
print("1. Ouvrir la webcam")
print("2. Utiliser une image")
choice = input("Entrez 1 ou 2 : ")

if choice == '1':
    # Option Webcam
    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        print("Erreur : Impossible d'ouvrir la webcam.")
        exit()

    while True:
        ret, frame = cap.read()
        if not ret:
            print("Erreur : Impossible de lire la frame depuis la webcam.")
            break

        # Détecter et afficher les émotions
        frame = detect_and_predict_emotion(frame)
        cv2.imshow('Webcam - Emotion Recognition', frame)

        # Quitter avec la touche 'q'
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

elif choice == '2':
    # Option Image
    print("Sélectionnez une image depuis vos fichiers...")
    image_path = easygui.fileopenbox(title="Choisissez une image", filetypes=["*.jpg", "*.jpeg", "*.png", "*.bmp", "*.tiff", "*.gif","*.HEIC"])

    if image_path:
        # Lire l'image sélectionnée
        image = cv2.imread(image_path)

        if image is None:
            print("Erreur : Impossible de charger l'image.")
        else:
            # Détecter et afficher les émotions sur l'image
            image = detect_and_predict_emotion(image)
            
            # Afficher l'image avec les visages et les émotions annotées
            cv2.imshow('Image - Emotion Recognition', image)
            cv2.waitKey(0)  # Attendre jusqu'à ce que l'utilisateur ferme la fenêtre
            cv2.destroyAllWindows()
    else:
        print("Aucune image sélectionnée.")

else:
    print("Option invalide. Veuillez choisir 1 ou 2.")
emotion_model = load_model(model_path)
emotion_labels = ['Colère', 'Dégoût', 'Peur', 'Joie', 'Tristesse', 'Surprise', 'Neutre']

# Charger le détecteur de visages Haar Cascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Fonction pour détecter les visages et prédire les émotions
def detect_and_predict_emotion(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)
    
    for (x, y, w, h) in faces:
        # Extraire et prétraiter le visage
        face = gray[y:y+h, x:x+w]
        face = cv2.resize(face, (48, 48))  # Redimensionner pour correspondre à l'entrée du modèle
        face = face / 255.0  # Normaliser les pixels entre 0 et 1
        face = np.expand_dims(face, axis=0)  # Ajouter la dimension batch
        face = np.expand_dims(face, axis=-1)  # Ajouter la dimension des canaux (1 pour grayscale)

        # Prédire l'émotion
        emotion_prediction = emotion_model.predict(face, verbose=0)
        emotion_label = emotion_labels[np.argmax(emotion_prediction)]

        # Annoter l'image avec l'émotion détectée
        cv2.putText(image, emotion_label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
        cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)

    return image

# Menu principal
print("Choisissez une option :")
print("1. Ouvrir la webcam")
print("2. Utiliser une image")
choice = input("Entrez 1 ou 2 : ")

if choice == '1':
    # Option Webcam
    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        print("Erreur : Impossible d'ouvrir la webcam.")
        exit()

    while True:
        ret, frame = cap.read()
        if not ret:
            print("Erreur : Impossible de lire la frame depuis la webcam.")
            break

        # Détecter et afficher les émotions
        frame = detect_and_predict_emotion(frame)
        cv2.imshow('Webcam - Emotion Recognition', frame)

        # Quitter avec la touche 'q'
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

elif choice == '2':
    # Option Image
    print("Sélectionnez une image depuis vos fichiers...")
    image_path = easygui.fileopenbox(title="Choisissez une image", filetypes=["*.jpg", "*.jpeg", "*.png", "*.bmp", "*.tiff", "*.gif","*.HEIC"])

    if image_path:
        # Lire l'image sélectionnée
        image = cv2.imread(image_path)

        if image is None:
            print("Erreur : Impossible de charger l'image.")
        else:
            # Détecter et afficher les émotions sur l'image
            image = detect_and_predict_emotion(image)
            
            # Afficher l'image avec les visages et les émotions annotées
            cv2.imshow('Image - Emotion Recognition', image)
            cv2.waitKey(0)  # Attendre jusqu'à ce que l'utilisateur ferme la fenêtre
            cv2.destroyAllWindows()
    else:
        print("Aucune image sélectionnée.")

else:
    print("Option invalide. Veuillez choisir 1 ou 2.")
```
